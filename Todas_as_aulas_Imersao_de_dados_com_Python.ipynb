```python
df_translated = df.rename(columns={
    'ano': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_empregado': 'residencia',
    'taxa_remoto': 'remoto',
    'localizacao_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
})

df = df_translated.copy()

print("Colunas traduzidas com sucesso. Aqui estão as 5 primeiras linhas do DataFrame atualizado:")
df.head()
```

```python
df["senioridade"].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

print("Categorias da coluna 'senioridade' traduzidas com sucesso. Aqui estão as novas contagens:")
df['senioridade'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

print("Categorias da coluna 'remoto' traduzidas com sucesso. Aqui estão as novas contagens:")
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

print("Categorias da coluna 'tamanho_empresa' traduzidas com sucesso. Aqui estão as novas contagens:")
df['tamanho_empresa'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})

print("Categorias da coluna 'contrato' traduzidas com sucesso. Aqui estão as novas contagens:")
df['contrato'].value_counts()
```

```python
import numpy as np

#criando um novo dataframe
df_salarios = pd.DataFrame({
  'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
  'salario': [4000, np.nan, 5000, np.nan, 100000]

})

#preenchendo os valores nulos com a media dos salarios e arredonda

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
df_salarios

#com a mediana

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffil"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfil"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "cidade": ["São Paulo", np.nan, "Curitiba", np.nan, "Belem"]

})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Nao informado")
df_cidades
```
