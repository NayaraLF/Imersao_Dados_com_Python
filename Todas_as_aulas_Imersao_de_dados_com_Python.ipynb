```python
df = df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_empregado',
    'remote_ratio': 'remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
})
```

```python
df["senioridade"].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
```

```python
import numpy as np

#criando um novo dataframe
df_salarios = pd.DataFrame({
  'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
  'salario': [4000, np.nan, 5000, np.nan, 100000]

})

#preenchendo os valores nulos com a media dos salarios e arredonda

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
df_salarios

#com a mediana

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffil"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfil"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "cidade": ["São Paulo", np.nan, "Curitiba", np.nan, "Belem"]

})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Nao informado")
df_cidades
```

Aula 03 - Visualizacao de dados

```python
df_limpo['senioridade'].value_counts().plot(kind='bar',title='Distribuicao de Senioridade')
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='senioridade', y='salario_em_usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='salario_em_usd')
plt.title('Salario medio por senioridade')
plt.xlabel('senioridade')
plt.ylabel('salario medio anual (USD)')
plt.show()
```

```python
df_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=False).index
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='salario_em_usd', order=ordem)
plt.title('Salario medio por senioridade')
plt.xlabel('senioridade')
plt.ylabel('salario medio anual (USD)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df_limpo['salario_em_usd'], bins=50, kde=True )
plt.title('Distribuicao dos salarios anuais')
plt.xlabel('Salario em USD')
plt.ylabel('Frequencia')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_limpo['salario_em_usd'])
plt.title('Boxplot salario')
plt.xlabel('Salario em USD')
plt.show()
```

```python
import plotly.express as px
```

```python
df_grouped_senioridade = df_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(
    df_grouped_senioridade,
    x='senioridade',
    y='salario_em_usd',
    title='Média Salarial por Senioridade (USD)',
    labels={'senioridade': 'Senioridade', 'salario_em_usd': 'Média Salarial (USD)'},
    color='senioridade'
)

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names = 'tipo_trabalho',
             values='quantidade',
             title='Proporcao dos tipos de trabalho',
             hole=0.5
             )
fig.update_traces(textinfo='percent+label')
fig.show()
```
